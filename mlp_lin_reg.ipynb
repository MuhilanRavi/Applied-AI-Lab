```python
import matplotlib.pyplot as plt

# Dataset: house size (sq.ft) and price ($1000s)
X = [650, 800, 1200, 1500, 2000]
Y = [75, 100, 150, 180, 250]

# Normalize inputs for stability
X = [x/2000 for x in X]   # scale between 0 and 1
Y = [y/250 for y in Y]    # scale between 0 and 1

# Parameters
w = 0.1   # weight
b = 0.0   # bias
lr = 0.05 # learning rate
threshold = 0.5  # user-defined threshold for step function

# Step activation function
def step_function(x, threshold=0.5):
    return 1 if x >= threshold else 0

# Training
epochs = 10
errors = []

for epoch in range(epochs):
    total_error = 0
    for x, y in zip(X, Y):
        # Forward pass
        net = w*x + b
        y_pred = step_function(net, threshold)

        # Error
        error = y - y_pred
        total_error += error**2

        # Weight update (Perceptron rule)
        w += lr * error * x
        b += lr * error

    errors.append(total_error)
    print(f"Epoch {epoch}: Total Error = {total_error:.2f}")

# Final parameters
print("\nTrained Parameters:")
print("Weight =", round(w,3), "Bias =", round(b,3))

# Visualization
plt.figure(figsize=(8,5))
plt.plot(range(epochs), errors, marker='o', color='blue')
plt.title("Error vs Epochs")
plt.xlabel("Epochs")
plt.ylabel("Total Error")
plt.grid(True)
plt.show()
```
