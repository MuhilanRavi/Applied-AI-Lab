```python
# INPUTS
x1, x2, x3, x4 = 0, 0, 1, 1
X = [x1, x2, x3, x4]

# Target Output
y = 1

# Learning rate
lr = 0.72

# INITIAL WEIGHTS
# Input → Hidden Layer 1
w15, w16 = 0.2, -0.3
w25, w26 = 0.4, 0.1
w35, w36 = -0.5, 0.2
w45, w46 = 0.1, 0.3

# Hidden Layer 1 → Hidden Layer 2
w57, w58 = 0.3, -0.2
w67, w68 = 0.2, 0.4

# Hidden Layer 2 → Output
w79, w89 = -0.4, 0.2

# Biases
b5, b6 = 0.1, -0.2
b7, b8 = 0.2, 0.1
b9 = -0.3
```

```python
# ACTIVATION FUNCTIONS
import math

def sigmoid(x):
    return 1 / (1 + math.exp(-x))

def relu(x):
    return max(0, x)

def leaky_relu(x, alpha=0.01):
    return x if x > 0 else alpha * x

def tanh(x):
    return math.tanh(x)

def swish(x):
    return x * sigmoid(x)
```

```python
# FORWARD PROPAGATION
for epoch in range(10):
    # ----- Hidden Layer 1 -----
    net5 = x1*w15 + x2*w25 + x3*w35 + x4*w45 + b5
    net6 = x1*w16 + x2*w26 + x3*w36 + x4*w46 + b6

    x5 = relu(net5)
    x6 = relu(net6)

# ----- Hidden Layer 2 -----
    net7 = x5*w57 + x6*w67 + b7
    net8 = x5*w58 + x6*w68 + b8

    x7 = relu(net7)
    x8 = relu(net8)

# ----- Output Layer -----
    net9 = x7*w79 + x8*w89 + b9

# Change activation here if needed
    x9 = sigmoid(net9)

# BACKWARD PROPAGATION

# Output layer delta
    error9 = y - x9
    print(f"Epoch {epoch}: Output x9 = {x9:.2f}, Error = {error9:.2f}")
    delta9 = error9 * x9 * (1 - x9)

# Hidden Layer 2 deltas
    delta7 = delta9 * w79 * (1 if net7 > 0 else 0)
    delta8 = delta9 * w89 * (1 if net8 > 0 else 0)

# Hidden Layer 1 deltas
    delta5 = (delta7*w57 + delta8*w58) * (1 if net5 > 0 else 0)
    delta6 = (delta7*w67 + delta8*w68) * (1 if net6 > 0 else 0)

# WEIGHT UPDATES

# Hidden Layer 2 → Output
    w79 += lr * delta9 * x7
    w89 += lr * delta9 * x8
    b9  += lr * delta9

# Hidden Layer 1 → Hidden Layer 2
    w57 += lr * delta7 * x5
    w67 += lr * delta7 * x6
    w58 += lr * delta8 * x5
    w68 += lr * delta8 * x6
    b7  += lr * delta7
    b8  += lr * delta8

# Input → Hidden Layer 1
    w15 += lr * delta5 * x1
    w25 += lr * delta5 * x2
    w35 += lr * delta5 * x3
    w45 += lr * delta5 * x4
    b5  += lr * delta5

    w16 += lr * delta6 * x1
    w26 += lr * delta6 * x2
    w36 += lr * delta6 * x3
    w46 += lr * delta6 * x4
    b6  += lr * delta6
```
