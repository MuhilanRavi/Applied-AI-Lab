```python
# Inputs
x1, x2, x3, x4 = 1, 1, 0, 1
X = [x1, x2, x3, x4]
# Output
y = 1
# Learning rate
lr = 0.1
# Weight
w15, w16 = 0.3, 0.1
w25, w26 = -0.2, 0.4
w35, w36 = 0.2, -0.3
w45, w46 = 0.1, 0.4
w57, w67 = -0.3, 0.2
# bias
b5, b6, b7 = 0.2, 0.1, -0.3

# Activation function
import math
def sigmoid(x):
  return 1 / (1 + math.exp(-x))

# FORWARD PROPAGATION
# Hidden Layer
net5 = x1*w15 + x2*w25 + x3*w35 + x4*w45 + b5
net6 = x1*w16 + x2*w26 + x3*w36 + x4*w46 + b6
x5 = sigmoid(net5)
x6 = sigmoid(net6)
# Output Layer
net7 = x5*w57 + x6*w67 + b7
x7 = sigmoid(net7)
print("Forward Pass Results")
print("x5 =", round(x5, 4))
print("x6 =", round(x6, 4))
print("x7 =", round(x7, 4))

# BACKWARD PROPAGATION
# Output Layer Error
error7 = y - x7
delta7 = error7 * x7 * (1 - x7)
# Hidden Layer Errors
delta5 = delta7 * w57 * x5 * (1 - x5)
delta6 = delta7 * w67 * x6 * (1 - x6)

# Weight and bias update
# Update Hidden → Output w57 = w57 + lr * delta7 * x5 w67 = w67 + lr * delta7 * x6 b7 = b7 + lr * delta7
# Update Input → Hidden (Neuron x5)
w15 = w15 + lr * delta5 * x1
w25 = w25 + lr * delta5 * x2
w35 = w35 + lr * delta5 * x3
w45 = w45 + lr * delta5 * x4
b5 = b5 + lr * delta5
# Update Input → Hidden (Neuron x6)
w16 = w16 + lr * delta6 * x1
w26 = w26 + lr * delta6 * x2
w36 = w36 + lr * delta6 * x3
w46 = w46 + lr * delta6 * x4
b6 = b6 + lr * delta6

print("\nUpdated Weights and Biases")
print("w57 =", round(w57,4), "w67 =", round(w67,4), "b7 =", round(b7,4))
print("w15 =", round(w15,4), "w25 =", round(w25,4),
"w35 =", round(w35,4), "w45 =", round(w45,4), "b5 =", round(b5,4))
print("w16 =", round(w16,4), "w26 =", round(w26,4),
"w36 =", round(w36,4), "w46 =", round(w46,4), "b6 =", round(b6,4))
```
